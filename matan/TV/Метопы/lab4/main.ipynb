## Метод градиентного спуска
##### Рабочий пример
<a href = "https://www.desmos.com/3d/23b2b7a336">График функции</a>

```python
from lab4 import gradMethods
from numpy import exp
x0 = -1
y0 = 1
e = 0.0001
def f(x,y):
    return x**2+y**2-x+6+exp(x)
def fx(x):
    return 2*x-1+exp(x)
def fy(y):
    return 2*y
grad = gradMethods(x0,y0,e,f,fx,fy)
try:
    gradAnswer = grad.gradDescent()
    gradAnswer.print_result()
except OverflowError as e:
    print("Too many iterations. Try another starting point: this variant went to infinity.")
```

##### Нерабочий пример (по варианту)

```python
x01 = 0
y01 = 3 #works only on y0 = 3
e1 = 0.0001
def f1(x,y):
    return x**2-y**2-4*x+6*y
def fx1(x):
    return 2*x-4
def fy1(y):
    return -2*y+6
grad1 = gradMethods(x01,y01,e1,f1,fx1,fy1)
try:
    gradAnswer1 = grad1.gradDescent()
    gradAnswer1.print_result()
except OverflowError as e:
    print("Too many iterations. Try another starting point: this variant went to infinity.")
```

## Метод наискорейшего спуска
##### Рабочий пример

```python
try:
    fasterGradAnswer = grad1.fastestDescent()
    fasterGradAnswer.print_result()
except OverflowError as e:
    print("Too many iterations. Try another starting point: this variant went to infinity.")
```
